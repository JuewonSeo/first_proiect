```python
import pandas as pd
parkDF = pd.read_excel('/content/drive/MyDrive/졸업논문/입차데이터_표준편차33.xlsx')
parkDF
```

```python
df = pd.DataFrame({'차량번호' : parkDF['차량번호'],
                   '평균입차간격' : parkDF['평균입차간격'],
                   '평균입차시간' : parkDF['평균입차시간'],
                   '입차표준편차' : parkDF['입차표준편차'],
                   '구분' : parkDF['구분'],
                   '패턴' : parkDF['패턴']})

df
```

```python
df = df.dropna(axis = 0)
df
```

```python
df2 = df.reset_index(drop = False)
df2 = df.reset_index(drop = True)
df = df2
df
```

```python
cols = list(df.columns)
data_x = df[cols[1:4]]
data_x
```

```python
data_x.to_excel('t-test.xlsx', index = False)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
data_scaled = sc.fit_transform(data_x)
scaled = pd.DataFrame(data_scaled)
scaled
```

```python
scaled[2] = scaled[2] * 0.01
scaled
```

```python
scaled['구분'] = df['구분']
scaled
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
cluster = linkage(y = scaled, method = 'complete', metric = 'euclidean')
cluster
cluster.shape
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (30, 10))
dendrogram(cluster, leaf_rotation = 90, leaf_font_size=6, color_threshold = 4, )
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

cut_tree = fcluster(cluster, t = 4, criterion = 'distance')
cut_tree

labels = scaled['구분']

df4 = pd.DataFrame({'pred' : cut_tree, 'labels' : labels})
df4

con_mat = pd.crosstab(df4['pred'], df4['labels'])
con_mat
```

```python
data_x['cluster'] = cut_tree
data_x
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
import seaborn as sns
plt.figure(figsize = (20, 10))
ax = sns.scatterplot(x = '평균입차시간', y = '평균입차간격', hue = 'cluster', data = data_x, palette = 'Set2')
plt.show()
```
